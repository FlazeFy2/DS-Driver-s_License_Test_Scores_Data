```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Drivers License Data.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Gender','Age Group','Race','Training','Reactions','Qualified']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Comparison Of {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
df = pd.read_csv("Drivers License Data.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Qualified per Training
df["Qualified"] = df["Qualified"].replace({
    "Yes": "Qualified",
    "No": "Not Qualified"
})
df["Training_Qualified"] = df["Training"] + " Training and " + df["Qualified"]

# Grouping
grouped = df["Training_Qualified"].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
grouped.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Of Training - Qualified")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
