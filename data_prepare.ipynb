```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Drivers License Data.csv")

# Data Understanding - Variable & Data Type
print(df.dtypes)
print("\n")
print(f"{len(df.columns)} Column")
```

```python
# Data Preparation - Handle Missing Value
df.isnull().sum()
```

```python
# Data Understanding - Descriptive Statistic
df.describe()
```

```python
# Data Preparation - Outliers
ratio_column = ["Signals", "Yield", "Speed Control", "Night Drive", "Road Signs", "Steer Control", "Mirror Usage", "Confidence", "Parking", "Theory Test"]

for dt in ratio_column:
    plt.boxplot(df[dt])
    plt.title(f'Boxplot of {dt}')
    plt.show()
```

```python
# Data Preparation - Invalid Value
ratio_column = ["Signals", "Yield", "Speed Control", "Night Drive", "Road Signs", "Steer Control", "Mirror Usage", "Confidence", "Parking", "Theory Test"]
for dt in ratio_column:
    invalid_value = df[df[dt] < 0]
    print(invalid_value.index)
```
